```python
# OSM data for Trento
location = "Trento, Territorio Val d'Adige, Provincia di Trento, Trentino-Alto Adige/SÃ¼dtirol, Italy"
G = ox.graph_from_place(location, network_type='bike')

# Convert the graph into a geopandas dataframe
tn_nodes, tn_edges = ox.graph_to_gdfs(G)

# List of major road types
major_roads = ['primary', 'primary_link', 'secondary', 'secondary_link', 'tertiary',
               'tertiary_link', 'trunk', 'trunk_link', 'residential', 'cycleway',
               'living_street', 'unclassified', 'motorway', 'motorway_link',
               'pedestrian', 'steps', 'track']

```

```python
filtered_edges = tn_edges[tn_edges['highway'].isin(major_roads)].copy()
filtered_edges.head()
```

```python
# Filter tn_edges to include only major roads
filtered_edges = tn_edges[tn_edges['highway'].isin(major_roads)].copy()

# Convert lists in the "name" column to tuples
filtered_edges.loc[:, 'name'] = filtered_edges['name'].apply(lambda x: tuple(x) if isinstance(x, list) else x)

# Get distinct values and their counts in the 'name' column
distinct_names_counts = filtered_edges['name'].value_counts()

# Print the distinct values and their counts
print(distinct_names_counts["Via al Desert"])
```
